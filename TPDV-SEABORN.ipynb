```python
pal=dict(female="red" , male="yellow")
s = sns.FacetGrid(df,hue="sex",palette=pal)
s.map(sns.scatterplot, "age", "fare", s=50, alpha=.7 , linewidth=.5) 
s.add_legend()
```

```python

s = sns.FacetGrid(df, col="embarked",  row="sex")
s.map_dataframe(sns.scatterplot, x="fare", y="body")
s.set_axis_labels("fare", "body")
s.add_legend()
```

```python
g = sns.PairGrid(df, vars=["age","fare"],hue="survived")
g.map(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(df, hue="survived")
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")

```
